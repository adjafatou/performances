CHARGEMNT DE DONNEES


```python

fichier_resultats = "../data/benchmark_results.csv"
donnees = pd.read_csv(fichier_resultats)

colonnes_correct = [col for col in donnees.columns if col.startswith('ia_correct_')]
colonnes_temps = [col for col in donnees.columns if col.startswith('temps_reponse_')]
modeles = [col.replace('ia_correct_', '').replace('_', ':') for col in colonnes_correct]

print(f"Dataset: {len(donnees)} questions")
print(f"ModÃ¨les: {modeles}")
print(f"Colonnes: {list(donnees.columns)}")

print(donnees.head())
```

PERFORMANCE DU MEILLEUR MODEL


```python
print("PERFORMANCES")

resultats_globaux = {}
for i, col in enumerate(colonnes_correct):
    modele = modeles[i]
    precision = (donnees[col].sum() / len(donnees)) * 100
    resultats_globaux[modele] = precision
    print(f"{modele}: {precision:.1f}%")

fig, ax = plt.subplots(1, 2, figsize=(15, 6))

bars = ax[0].bar(resultats_globaux.keys(), resultats_globaux.values(), 
                color=['skyblue', 'lightcoral', 'lightgreen', 'gold'][:len(modeles)], 
                alpha=0.8)
ax[0].set_title('PrÃ©cision globale par modÃ¨le', fontsize=14)
ax[0].set_ylabel('PrÃ©cision (%)')
ax[0].set_ylim(0, 100)

for i, (modele, precision) in enumerate(resultats_globaux.items()):
    ax[0].text(i, precision + 2, f'{precision:.1f}%', ha='center')

if len(modeles) > 0:
    correct_count = donnees[colonnes_correct[0]].sum()
    incorrect_count = len(donnees) - correct_count
    
    ax[1].pie([correct_count, incorrect_count], 
              labels=['Correctes', 'Incorrectes'], 
              autopct='%1.1f%%', 
              colors=['lightgreen', 'lightcoral'])
    ax[1].set_title(f'Distribution - {modeles[0]}', fontsize=14)

plt.tight_layout()
plt.show()

print(f"Meilleur modÃ¨le: {max(resultats_globaux, key=resultats_globaux.get)} ({max(resultats_globaux.values()):.1f}%)")
```

ANALYSE POUR LES CATEGORIES

```python
if 'categorie' in donnees.columns:
    print("ANALYSE PAR CATÃ‰GORIE")
    
    fig, axes = plt.subplots(1, len(modeles), figsize=(6*len(modeles), 6))
    if len(modeles) == 1:
        axes = [axes]
    
    tableau_categories = pd.DataFrame()
    
    for i, (col_correct, modele) in enumerate(zip(colonnes_correct, modeles)):
        precision_categorie = donnees.groupby('categorie')[col_correct].agg(['mean', 'count'])
        precision_categorie['precision'] = precision_categorie['mean'] * 100
        
        tableau_categories[modele] = precision_categorie['precision']
        
        precision_categorie['precision'].plot(kind='bar', ax=axes[i], 
                                             color='lightcoral', alpha=0.8)
        axes[i].set_title(f'PrÃ©cision par catÃ©gorie - {modele}')
        axes[i].set_ylabel('PrÃ©cision (%)')
        axes[i].set_ylim(0, 100)
        axes[i].tick_params(axis='x', rotation=45)
        
        for j, (cat, prec) in enumerate(precision_categorie['precision'].items()):
            axes[i].text(j, prec + 2, f'{prec:.1f}%', ha='center')
    
    plt.tight_layout()
    plt.show()
    
    print("Tableau de prÃ©cision par catÃ©gorie:")
    print(tableau_categories.round(1))
    
    print("Statistiques par catÃ©gorie:")
    for cat in donnees['categorie'].unique():
        nb_questions = len(donnees[donnees['categorie'] == cat])
        print(f"   {cat}: {nb_questions} questions")

else:
    print("Colonne 'categorie' non trouvÃ©e")
```

ANALSE POUR LES DIFFICULTE

```python
if 'difficulte' in donnees.columns:
    print("ANALYSE PAR DIFFICULTÃ‰")
    
    tableau_difficulte = pd.DataFrame()
    for col_correct, modele in zip(colonnes_correct, modeles):
        precision_diff = donnees.groupby('difficulte')[col_correct].mean() * 100
        tableau_difficulte[modele] = precision_diff
    
    fig, ax = plt.subplots(1, 1, figsize=(12, 6))
    sns.heatmap(tableau_difficulte.T, 
                annot=True, 
                fmt='.1f', 
                cmap='RdYlGn', 
                center=50, 
                ax=ax, 
                cbar_kws={'label': 'PrÃ©cision (%)'})
    
    ax.set_title('PrÃ©cision par niveau de difficultÃ©', fontsize=14)
    ax.set_xlabel('Niveau de difficultÃ©')
    ax.set_ylabel('ModÃ¨les')
    
    plt.tight_layout()
    plt.show()
    
    print("Tableau de prÃ©cision par difficultÃ©:")
    print(tableau_difficulte.round(1))
    
    print("RÃ©partition des questions par difficultÃ©:")
    for diff in donnees['difficulte'].value_counts().index:
        nb_questions = len(donnees[donnees['difficulte'] == diff])
        print(f"   {diff}: {nb_questions} questions")

else:
    print("Colonne 'difficulte' non trouvÃ©e")
```

COMPARAISON NTRE LES TEMPS DE REPONSES

```python
if colonnes_temps:
    print("ANALYSE DES TEMPS DE RÃ‰PONSE")
    
    stats_temps = {}
    for col_temps, modele in zip(colonnes_temps, modeles):
        temps = donnees[col_temps]
        stats_temps[modele] = {
            'moyenne': temps.mean(),
            'mediane': temps.median(),
            'min': temps.min(),
            'max': temps.max(),
            'ecart_type': temps.std()
        }
    
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    axes = axes.flatten()
    
    for i, (col_temps, modele) in enumerate(zip(colonnes_temps, modeles)):
        if i < len(axes) - 1:
            donnees[col_temps].hist(bins=20, ax=axes[i], alpha=0.7, color=f'C{i}')
            axes[i].set_title(f'Distribution des temps - {modele}')
            axes[i].set_xlabel('Temps (secondes)')
            axes[i].set_ylabel('FrÃ©quence')
            
            moyenne = stats_temps[modele]['moyenne']
            axes[i].axvline(moyenne, color='red', linestyle='--', 
                           label=f'Moyenne: {moyenne:.2f}s')
            axes[i].legend()
    
    if len(modeles) > 1:
        temps_data = [donnees[col] for col in colonnes_temps]
        bp = axes[-1].boxplot(temps_data, labels=modeles, patch_artist=True)
        
        colors = ['lightblue', 'lightcoral', 'lightgreen', 'gold']
        for patch, color in zip(bp['boxes'], colors):
            patch.set_facecolor(color)
            patch.set_alpha(0.7)
            
        axes[-1].set_title('Comparaison des temps de rÃ©ponse')
        axes[-1].set_ylabel('Temps (secondes)')
        axes[-1].tick_params(axis='x', rotation=45)
    else:
        axes[-1].remove()
    
    plt.tight_layout()
    plt.show()
    
    df_stats_temps = pd.DataFrame(stats_temps).T
    print("Statistiques des temps de rÃ©ponse (secondes):")
    print(df_stats_temps.round(3))
    
    print("CorrÃ©lations temps vs prÃ©cision:")
    for col_correct, col_temps, modele in zip(colonnes_correct, colonnes_temps, modeles):
        correlation = donnees[col_correct].astype(int).corr(donnees[col_temps])
        print(f"   {modele}: {correlation:.3f}")
    
    modele_rapide = min(stats_temps, key=lambda x: stats_temps[x]['moyenne'])
    temps_rapide = stats_temps[modele_rapide]['moyenne']
    print(f"ModÃ¨le le plus rapide: {modele_rapide} ({temps_rapide:.2f}s)")

else:
    print("Aucune colonne de temps trouvÃ©e")
```

ANALYSE DES QUESTIONS

```python
print("ANALYSE DES QUESTIONS")

if colonnes_correct:
    taux_echec = 1 - donnees[colonnes_correct].mean(axis=1)
    taux_reussite = donnees[colonnes_correct].mean(axis=1)
    
    print("TOP 5 QUESTIONS LES PLUS DIFFICILES")
    questions_difficiles = donnees.loc[taux_echec.nlargest(5).index]
    
    for idx, (_, row) in enumerate(questions_difficiles.iterrows()):
        echec_pct = taux_echec.loc[row.name] * 100
        question_courte = row['question'][:70] + "..." if len(row['question']) > 70 else row['question']
        print(f"{idx+1}. {question_courte}")
        print(f"   Taux d'Ã©chec: {echec_pct:.1f}%")
        print(f"   RÃ©ponse: {row['reponse_correcte']}")
        print()
    
    print("TOP 5 QUESTIONS LES PLUS FACILES")
    questions_faciles = donnees.loc[taux_reussite.nlargest(5).index]
    
    for idx, (_, row) in enumerate(questions_faciles.iterrows()):
        reussite_pct = taux_reussite.loc[row.name] * 100
        question_courte = row['question'][:70] + "..." if len(row['question']) > 70 else row['question']
        print(f"{idx+1}. {question_courte}")
        print(f"   Taux de rÃ©ussite: {reussite_pct:.1f}%")
        print(f"   RÃ©ponse: {row['reponse_correcte']}")
        print()
    
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))
    
    (taux_reussite * 100).hist(bins=20, ax=axes[0], alpha=0.7, color='skyblue')
    axes[0].set_title('Distribution des taux de rÃ©ussite par question')
    axes[0].set_xlabel('Taux de rÃ©ussite (%)')
    axes[0].set_ylabel('Nombre de questions')
    axes[0].axvline((taux_reussite * 100).mean(), color='red', linestyle='--',
                   label=f'Moyenne: {(taux_reussite * 100).mean():.1f}%')
    axes[0].legend()
    
    def categoriser_difficulte(taux):
        if taux >= 0.8:
            return "TrÃ¨s facile (â‰¥80%)"
        elif taux >= 0.6:
            return "Facile (60-79%)"
        elif taux >= 0.4:
            return "Moyen (40-59%)"
        elif taux >= 0.2:
            return "Difficile (20-39%)"
        else:
            return "TrÃ¨s difficile (<20%)"
    
    difficulte_auto = taux_reussite.apply(categoriser_difficulte)
    difficulte_counts = difficulte_auto.value_counts()
    
    colors = ['darkred', 'red', 'orange', 'lightgreen', 'green']
    axes[1].pie(difficulte_counts.values, 
               labels=difficulte_counts.index,
               autopct='%1.1f%%',
               colors=colors[:len(difficulte_counts)])
    axes[1].set_title('RÃ©partition automatique par difficultÃ©')
    
    plt.tight_layout()
    plt.show()
    
    print("STATISTIQUES GLOBALES")
    print(f"Questions trÃ¨s faciles (â‰¥80%): {(taux_reussite >= 0.8).sum()}")
    print(f"Questions faciles (60-79%): {((taux_reussite >= 0.6) & (taux_reussite < 0.8)).sum()}")
    print(f"Questions moyennes (40-59%): {((taux_reussite >= 0.4) & (taux_reussite < 0.6)).sum()}")
    print(f"Questions difficiles (20-39%): {((taux_reussite >= 0.2) & (taux_reussite < 0.4)).sum()}")
    print(f"Questions trÃ¨s difficiles (<20%): {(taux_reussite < 0.2).sum()}")

else:
    print("Aucune colonne de rÃ©sultats trouvÃ©e")
```
